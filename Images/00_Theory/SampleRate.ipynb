```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
from scipy.interpolate import interp1d
```

```python
X = np.arange(0,20,0.01)
Y = np.cos(X)

X1 = np.arange(-1,22,1.5)
X2 = np.arange(-1.4,20,3)
Y1 = np.cos(X1)
Y2 = np.cos(X2)

n1, = X1.shape
n2, = X2.shape

X_Y_Spline = make_interp_spline(X2, Y2)
 
# Returns evenly spaced numbers
# over a specified interval.
X_ = np.linspace(X2.min(), X2.max(), 500)
Y_ = X_Y_Spline(X_)


X_Y_Spline2 =interp1d(X1, Y1, kind = "cubic")
 
# Returns evenly spaced numbers
# over a specified interval.
X1_ = np.linspace(X1.min(), X1.max(), 500)
Y1_ = X_Y_Spline2(X1_)
```

```python
plt.figure(figsize=(10,5))

plt.plot(X,Y, label="Raw Analog Signal", c='k')
plt.plot(X1_,Y1_, "--b", label = "Signal with sample interval 1.5TU")
plt.plot(X_,Y_, "--r", label = "Signal with sample interval 3TU")



plt.scatter(X1,Y1,c='b', label = "Sample interval 1.5 TU")
plt.scatter(X2,Y2, c="r", label = "Sample interval 3 TU")


for i in range(n1):
    plt.plot([X1[i], X1[i]],[0,Y1[i]], 'b')
    
for i in range(n2):
    plt.plot([X2[i], X2[i]],[0,Y2[i]], 'r')

    


plt.plot([0,20], [0,0], "k")
plt.xlim(0,20)
plt.xlabel("Time (TU)")
plt.ylabel("Signal Amplitude")
plt.legend(loc=4)
plt.savefig("./SampleRate.png")
```
